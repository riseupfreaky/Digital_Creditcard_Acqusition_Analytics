# Creating a Blank CSV File with Labelled Columns

In this lesson, we'll learn how to create a blank CSV file with column headers using Python. This is a fundamental skill for any data analyst or programmer!

## 1. Import Required Libraries

First, we need to import the `csv` module, which is a built-in Python module that allows us to work with CSV files.

```python
# Import the csv module - a built-in Python module for working with CSV files
import csv
```

## 2. Define File Path and Column Headers

Now we'll specify where to save the file and what columns we want. Think of column headers as the labels at the top of each column - like "Name", "Age", "City", etc.

```python
# Define the file path where we want to save the CSV
# You can change this to any location on your computer
file_path = r"c:\Users\freak\.cache\kagglehub\datasets\student_data.csv"

# Define the column headers (labels for each column)
# These are like the column titles in a spreadsheet
column_headers = ["Name", "Age", "Grade", "City", "Phone"]
```

## 3. Create the Blank CSV File

Now comes the magic! We'll use the `csv.writer()` function to write our column headers to a new file. This creates a blank CSV with just the header row.

```python
# Open the file in write mode ('w') and create a CSV writer object
with open(file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write the column headers as the first row
    writer.writerow(column_headers)
    
print(f"✅ Blank CSV file created successfully at: {file_path}")
```

## 4. Verify the File Was Created

Let's read and display the contents of the newly created CSV file to confirm everything worked correctly!

```python
# Read and display the contents of the created CSV file
print("📄 Contents of the CSV file:")
print("-" * 40)

with open(file_path, mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)

print("-" * 40)
print("🎉 Great job! You've created a blank CSV file with labelled columns!")
```
